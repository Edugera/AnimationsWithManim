## Posicionamento Mobjects 

A seguir, vamos examinar algumas técnicas básicas para posicionar *`Mobjects`*.

Abra `scene.pye` adicione o seguinte *snippet* de código abaixo do SquareToCirclemétodo:

```python
#Renderize SquareAndCircle executando o seguinte comando na linha de comando
%%manim -qm -v WARNING SquareAndCircle

class SquareAndCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_fill(PINK, opacity=0.5)  # set the color and transparency

        square = Square()  # create a square
        square.set_fill(BLUE, opacity=0.5)  # set the color and transparency

        square.next_to(circle, RIGHT, buff=0.5)  # set the position
        self.play(Create(circle), Create(square))  # show the shapes on screen
```

`next_to` é um **Mobject** método para posicionar **Mobjects**.

Especificamos primeiro o círculo rosa como o ponto de referência do quadrado, passando circlecomo o primeiro argumento do método. O segundo argumento é usado para especificar a direção em que Mobjecté colocado em relação ao ponto de referência. Nesse caso, definimos a direção como RIGHT, dizendo a Manim para posicionar o quadrado à direita do círculo. Por fim, `buff=0.5` aplicou-se um pequeno buffer de distância entre os dois objetos.

Tente mudar RIGHT para LEFT, UP ou, em DOWN vez disso, e veja como isso muda a posição do quadrado.

Usando métodos de posicionamento, você pode renderizar uma cena com vários Mobjects, definindo suas localizações na cena usando coordenadas ou posicionando-as em relação umas às outras.

Para obter mais informações sobre next_toe outros métodos de posicionamento, verifique a lista de Mobjectmétodos em nosso manual de referência.

## Usando .animatesintaxe para animar métodos

A lição final neste tutorial é usar .animateum Mobjectmétodo que anima as alterações feitas em a Mobject. Quando você precede .animatequalquer chamada de método que modifica a Mobject, o método se torna uma animação que pode ser reproduzida usando self.play. Vamos voltar a SquareToCirclepara ver as diferenças entre usar métodos ao criar um Mobjecte animar essas chamadas de método com .animate.

Abra scene.pye adicione o seguinte snippet de código abaixo da SquareAndCircleclasse:

```python
# Renderize AnimatedSquareToCircleexecutando o seguinte comando na linha de comando
%%manim -qm -v WARNING AnimatedSquareToCircle

class AnimatedSquareToCircle(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        square = Square()  # create a square

        self.play(Create(square))  # show the shapes on screen
        self.play(square.animate.rotate(PI / 4))  # rotate the square
        self.play(
            ReplacementTransform(square, circle)
        )  # transform the square into a circle
        self.play(
            circle.animate.set_fill(PINK, opacity=0.5)
        )  # color the circle on screen
```

O primeiro self.playcria o quadrado. O segundo anima girando-o 45 graus. O terceiro transforma o quadrado em um círculo e o último colore o círculo em rosa. Embora o resultado final seja o mesmo de SquareToCircle, .animatemostra rotatee set_fillé aplicado ao Mobjectdinamicamente, em vez de criá-los com as alterações já aplicadas.

Experimente outros métodos, como flipou shift, e veja o que acontece.

Abra scene.pye adicione o seguinte snippet de código abaixo da AnimatedSquareToCircleclasse:

```python
# Renderize DifferentRotationsexecutando o seguinte comando na linha de comando
%%manim -qm -v WARNING DifferentRotations


class DifferentRotations(Scene):
    def construct(self):
        left_square = Square(color=BLUE, fill_opacity=0.7).shift(2 * LEFT)
        right_square = Square(color=GREEN, fill_opacity=0.7).shift(2 * RIGHT)
        self.play(
            left_square.animate.rotate(PI), Rotate(right_square, angle=PI), run_time=2
        )
        self.wait()
```

Isso Sceneilustra as peculiaridades de .animate. Ao usar .animate, Manim realmente pega um Mobjectestado inicial e seu estado final e interpola os dois. Na AnimatedSquareToCircleaula, você pode observar isso quando o quadrado gira: os cantos do quadrado parecem se contrair ligeiramente à medida que se movem para as posições necessárias para que o primeiro quadrado se transforme no segundo.

Em DifferentRotations, a diferença entre .animatea interpretação de rotação e o Rotatemétodo é muito mais aparente. Os estados inicial e final de uma Mobjectrotação de 360 ​​graus são os mesmos, então .animatetenta interpolar dois objetos idênticos e o resultado é o quadrado esquerdo. Se você descobrir que seu próprio uso de .animateestá causando um comportamento indesejado semelhante, considere o uso de métodos convencionais de animação, como o quadrado certo, que usa Rotate.
